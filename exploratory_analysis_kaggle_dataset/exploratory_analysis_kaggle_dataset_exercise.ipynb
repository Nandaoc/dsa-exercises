```python
# Importando bibliotecas
import os
import subprocess
import stat
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from datetime import datetime
sns.set(style = "white")
%matplotlib inline
```

## Distribuição de Veículos com base no Ano de Registro

### Exercício 1

Crie um Plot com a Distribuição de Veículos com base no Ano de Registro.

```python
# referência: https://stackoverflow.com/questions/14770735/how-do-i-change-the-figure-size-with-subplots
fig, ax = plt.subplots(figsize = (8,6))
plt.hist(df['yearOfRegistration'])
plt.ylabel("Densidade (KDE)")
plt.xlabel("Ano de Registro")
plt.title('Distribuição de Veículos com Base no Ano de Registro', fontdict = {'size': 14})
plt.show()
```

## Variação da faixa de preço pelo tipo de veículo

### Exercício 2

Crie um Boxplot para avaliar os outliers.

```python
fig, ax = plt.subplots(figsize = (10,8))
sns.boxplot(x = 'vehicleType', y = 'price', data = df)
ax.xaxis.set_label_text('Tipo de Veículo',fontdict = {'size':14})
ax.yaxis.set_label_text('Range de Preço',fontdict = {'size':14})
ax.set_title('Variação da Faixa de Preço pelo Tipo de Veículo', fontdict = {'size': 16})
plt.show()
```

## Contagem total de vaículos à venda conforme tipo de veículo

### Exercício 3

Crie um `count plot` que mostre o número de vículos pertencentes a cada categoria

```python
# referência: https://seaborn.pydata.org/generated/seaborn.countplot.html
fig, ax = plt.subplots(figsize = (10,8))
sns.countplot(x = 'vehicleType', data = df)
ax.xaxis.set_label_text('Tipo de Veículo',fontdict = {'size':16})
ax.yaxis.set_label_text('Total de Veículos Para Venda', fontdict = {'size':16})
ax.set_title('Contagem total de Veículos à venda Conforme o Tipo', fontdict = {'size': 16})
plt.show()
```

## Número de veículos pertencentes a cada marca

### Exercício 4

Crie um Plot que mostre o número de veículos pertencentes a cada marca.

```python
fig, ax = plt.subplots(figsize = (10,8))
sns.countplot(y = 'brand', data = df)
ax.set_title('Número de Veículos Pertencentes a Cada Marca',fontdict = {'size':18})
ax.xaxis.set_label_text('Número de Veículos',fontdict = {'size':16})
ax.yaxis.set_label_text('Marca',fontdict = {'size':16})
plt.show()
```

## Preço médio dos veículos com base no tipo de veículo, bem como no tipo de caixa de câmbio

### Exercício 5 

Crie um Plot com o Preço médio dos veículos com base no tipo de veículo, bem como no tipo de caixa de câmbio.

```python
fig, ax = plt.subplots(figsize = (10,8))
sns.barplot(x = 'vehicleType', y = 'price', hue = 'gearbox', data = df)
ax.set_title('Preço Médio por Tipo de Vaículo e Caixa de Câmbio',fontdict = {'size':18})
ax.xaxis.set_label_text('Tipo de Veículo',fontdict = {'size':16})
ax.yaxis.set_label_text('Preço',fontdict = {'size':16})
plt.show()
```
